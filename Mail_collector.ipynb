```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
```

```python
df=pd.read_excel('Ecb.xlsx', error_bad_lines=False, header=0)#прочитать список веб-адресов из файла Excel
df.head()#посмотреть датафрейм
df['Контакты']="Не определен"
```

```python
listt=0
for i, items in enumerate(df['Site'], 0):
    if i>0: #Если код останавливается на какой-то строке, написать последний номер, выданный в окне ниже, вместо 0 (это необработанная ошибка, её проще вручную обойти, чем разбираться, почему сайт не работает)
        if df['Контакты'][i]=='Не определен':
            try:
                print(i)
                url=items+"/contacts"
                if items.find('http')<0:
                    url='http://'+url
                Wiki=requests.get(url)
                a=bs(Wiki.text, 'html.parser')
                if str(a).find('mailto')!=-1:
                    l=str(a).find('mailto')+7
                    r=str(a).find('mailto')+50
                    a=(str(a)[l:r])
                    df['Контакты'][i]=a[:a.find('"')]
            except requests.exceptions.ConnectionError:
                print(i)
                listt+=1
```
